```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
beta_dist = np.random.beta(5, 10, 100)

beta_mean = beta_dist.mean()
beta_std = beta_dist.std()

print('The mean of this Beta distribution is {}'.format(round(beta_mean, 2)))
print('The standard deviation of this Beta distribution is {}'.format(round(beta_std, 2)))

plt.hist(beta_dist)
plt.title('Beta Distribution', fontsize=18)
plt.axvline(beta_dist.mean(), color='k', linewidth=2)
plt.axvline(beta_dist.mean() + beta_dist.std(), color='gray', linewidth=2)
plt.axvline(beta_dist.mean() - beta_dist.std(), color='gray', linewidth=2)
plt.show()
```

The mean seems to be a reasonable measure of central tendancy for this beta distribution, though it does seem to be slightly right skewed.  This skewness probably means that the standard deviation doesn't accurately describe the data.

```python
exp_dist = np.random.exponential(5, 100)

exp_mean = exp_dist.mean()
exp_std = exp_dist.std()

print('The mean of this Exponential distribution is {}'.format(round(exp_mean, 2)))
print('The standard deviation of this Exponential distribution is {}'.format(round(exp_std, 2)))

plt.hist(exp_dist)
plt.title('Exponential Distribution', fontsize=18)
plt.axvline(exp_dist.mean(), color='k', linewidth=2)
plt.axvline(exp_dist.mean() + exp_dist.std(), color='gray', linewidth=2)
plt.axvline(exp_dist.mean() - exp_dist.std(), color='gray', linewidth=2)
plt.show()
```

This is not a normal distribution.  The descriptive stats provide no insight.

```python
cauchy_dist = np.random.standard_cauchy(100)

cauchy_mean = cauchy_dist.mean()
cauchy_std = cauchy_dist.std()

print('The mean of this Standard Cauchy distribution is {}'.format(round(cauchy_mean, 2)))
print('The standard deviation of this Standard Cauchy distribution is {}'.format(round(cauchy_std, 2)))

plt.hist(cauchy_dist, bins=30)
plt.title('Standard Cauchy Distribution', fontsize=18)
plt.axvline(cauchy_dist.mean(), color='k', linewidth=2)
plt.axvline(cauchy_dist.mean() + cauchy_dist.std(), color='gray', linewidth=2)
plt.axvline(cauchy_dist.mean() - cauchy_dist.std(), color='gray', linewidth=2)
plt.show()
```

This is a very centered distribution with very short tails and some extreme outliers.  The mean does a good job but the standard deviation is greatly affected by the outliers.

```python
tri_dist = np.random.triangular(-10, 0, 10, 1000)

tri_mean = tri_dist.mean()
tri_std = tri_dist.std()

print('The mean of this Triangular distribution is {}'.format(round(tri_mean, 2)))
print('The standard deviation of this Triangular distribution is {}'.format(round(tri_std, 2)))

plt.hist(tri_dist, bins=20)
plt.title('Triangular Distribution', fontsize=18)
plt.axvline(tri_dist.mean(), color='k', linewidth=2)
plt.axvline(tri_dist.mean() + tri_dist.std(), color='gray', linewidth=2)
plt.axvline(tri_dist.mean() - tri_dist.std(), color='gray', linewidth=2)
plt.show()
```

This Triangular distribution is sort of normal, with a good mean central tendancy.  It has heavy tails which pulls the standard deviation away from the mean.

```python
# Additional Drill task of creating 2 normally-distributed variable and adding them together

norm1 = np.random.normal(5, 0.5, 1000)
norm2 = np.random.normal(10, 1, 1000)
norm3 = norm1 + norm2

norm3_mean = norm3.mean()
norm3_std = norm3.std()

print('The mean of the norm3 distribution is {}'.format(round(norm3_mean, 2)))
print('The standard deviation of the norm3 distribution is {}'.format(round(norm3_std, 2)))

plt.hist(norm3, bins=20)
#plt.hist(norm1, color='red', alpha=0.5)
#plt.hist(norm2, color='green', alpha=0.5)
plt.title('norm3 Distribution', fontsize=18)
plt.axvline(norm3.mean(), color='k', linewidth=2)
plt.axvline(norm3.mean() + norm3.std(), color='gray', linewidth=2)
plt.axvline(norm3.mean() + 2 * norm3.std(), color='gray', linestyle='dashed', linewidth=2)
plt.axvline(norm3.mean() - norm3.std(), color='gray', linewidth=2)
plt.axvline(norm3.mean() - 2* norm3.std(), color='gray', linestyle='dashed', linewidth=2)
plt.show()
```

Visually, this is a normal distribution.  The mean is essentially equal to the summed means from norm1 and norm2.  Interestingly though, the standard deviation is larger than either of the parent distributions, where intuitively one might think that the resulting standard deviation would be between the larger and smaller parent standard deviations.
