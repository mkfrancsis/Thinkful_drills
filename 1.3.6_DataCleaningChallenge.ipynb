```python
df['journal'] = journals
```

```python
topj = df[(df['journal'] == 'plos one')
          |(df['journal'] == 'journal of biological chemistry')
          |(df['journal'] == 'neuroimage')
          |(df['journal'] == 'plos genetics')
          |(df['journal'] == 'plos pathogens')
         ]
topj
```

```python
ster = topj['cost'].replace({'\$': '', '£': ''}, regex=True)
ster = ster.astype('float16')
ster = ster.replace(np.inf, 2504)
```

```python
topj = topj.assign(sterling=ster.values)
```

```python
top_journals = pd.Series(['journal of biological chemistry',
                          'neuroimage',
                          'plos genetics',
                          'plos one',
                          'plos pathogens'
                         ]
                        )
article_count = topj.groupby('journal')['article'].count()
mean_price = topj.groupby('journal')['sterling'].mean()
median_price = topj.groupby('journal')['sterling'].median()
```

```python
agg_stats = pd.DataFrame()
agg_stats = agg_stats.assign(journals=top_journals.values)
agg_stats = agg_stats.assign(num_articles=article_count.values)
agg_stats = agg_stats.assign(mean_sub_price=mean_price.values)
agg_stats = agg_stats.assign(median_sub_price=median_price.values)
agg_stats
```

```python
agg_stats['mean_art_price'] = agg_stats['mean_sub_price'] / agg_stats['num_articles']
agg_stats['median_art_price'] = agg_stats['median_sub_price'] / agg_stats['num_articles']
```
